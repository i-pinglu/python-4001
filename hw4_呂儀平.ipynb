```python
# 1. 使用 numpy package，使用者輸入 A 矩陣 (2,3) 以及 B 矩陣 (3,2)，使用矩陣乘法算出最後的結果 (30 分)
import numpy as np

A_row0 = input("A:\n")
A_row1 = input()
B_row0 = input("B:\n")
B_row1 = input()
B_row2 = input()

A = np.array([list(map(int, A_row0.split())), list(map(int, A_row1.split()))])
B = np.array([list(map(int, B_row0.split())), list(map(int, B_row1.split())), list(map(int, B_row2.split()))])
ans = np.matmul(A,B)
print(ans)
```

```python
# 2. 學校有 3 個班級，每個班級分別 10 人，請協助 Thousand 使用 numpy package 將 10 人的總和算出，以及找出每個班級最高分 (30 分)
import numpy as np

A = input("A class: ")
B = input("B class: ")
C = input("C class: ")

A = np.array(list(map(int, A.split())))
print(f"A class 總和: {np.sum(A, axis=0):.1f}, 最大: {np.max(A,axis=0):.1f}")
B = np.array(list(map(int, B.split())))
print(f"B class 總和: {np.sum(B, axis=0):.1f}, 最大: {np.max(B,axis=0):.1f}")
C = np.array(list(map(int, C.split())))
print(f"C class 總和: {np.sum(C, axis=0):.1f}, 最大: {np.max(C,axis=0):.1f}")
```
