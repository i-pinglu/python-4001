```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/i-pinglu/python-4001/main/Student%20Study%20data.csv")
```

每天不是在睡覺跟讀書的時間

```python
df['not sleeping or study'] = 24 - df['study hour'] - df['your sleep hour?']
df.head(5)
```

性別比例

```python
pie_chart = df['Your gender?'].value_counts().plot.pie(labels = ["Boys", "Girls"], autopct = '%1.1f%%', fontsize = 8, radius = 1.2)
plt.title("Gender")
plt.show()
```

每個人讀書時間超過3天的天數

```python
names = set(df["Name"])
days = []
for name in names:
  days.append(df.loc[(df["Name"] == name) & (df["study hour"] >= 3)].shape[0])
```

```python
from matplotlib import figure
import numpy as np

x = np.arange(len(names))
plt.plot(x, days)
plt.xticks(x, names)
plt.title('study time >= 3')
plt.xlabel("names")
plt.ylabel("days")
plt.show()
```

每天大家念書總時數

```python
df.loc[df["Day"] == "Monday"]["study hour"].values
```

```python
weeks = set(df["Day"])
x = np.arange(len(weeks))
width = 0.3
hours = []
for week in weeks:
  hours.append(df.loc[df["Day"] == week]["study hour"].sum(axis=0))
plt.figure(figsize=(11,5))
plt.bar(x, hours) 
plt.xticks(x, weeks)
plt.ylabel('Math')
plt.title('Final Term')
plt.show()
```
