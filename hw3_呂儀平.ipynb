```python
# Thousand 想要設計一個函式獲得兩點間的距離，請將程式碼完成 (20 分)
# A. 函式名稱為 get_distance
# B. 傳入的參數為 point1, point2 (參數是 list)
# C. point1 = [3,5], point2 = [7,2] (舉例)
# D. 距離公式為
# E. 回傳答案回主程式

def get_distance(point1, point2):
 x_sum = (point1[0] - point2[0])**2
 y_sum = (point1[1] - point2[1])**2
 sum = (x_sum + y_sum)**(1/2)
 return sum

point1 = input().split(" ")
for i in range(len(point1)):
   point1[i] = int(point1[i])

point2 = input().split(" ")
for i in range(len(point2)):
   point2[i] = int(point2[i])

ans = get_distance(point1 = point1, point2 = point2)
print("{:.2f}".format(ans))
```

```python
# 使用者輸入兩個數字，第一個是 row，第二個是 column，產生一個 matrix為 row x col，內部數值使用 random 函數，範圍定在 1 – 20 之間的整數。(25 分)
import random
row = int(input("row: "))
col = int(input("col: "))
matrix = [[0 for j in range(col)] for i in range(row)]
for i in range(row):
  for j in range(col):
    matrix[i][j] = random.randint(1,20)
    print(matrix[i][j],end=" ")
  print()
```
